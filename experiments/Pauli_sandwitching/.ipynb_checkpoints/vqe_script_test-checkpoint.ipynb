```python
Coeff = {}
Weight = []
Measured_List = []
Hamiltonian_Op = []
List_ops = []
Dict = {"paulis":[]}
f = open('QuHamiltonian.txt',"r") # Here we read in the qubit Hamiltonian from the .txt file

for line in f:
    a = line.strip().split('    ')
    b = {str(a[0]): float(a[1])}
    c = (float(a[1]), str(a[0]))
    if 'X' in str(a[0]) or 'Y' in str(a[0]):
        Hamiltonian_Op.append(c)
        Weight.append(float(a[1]))
        Coeff.update(b)
        
    Measured_List.append(str(a[0]))
f.close()
```

```python
for i in range(len(Weight)):
    a = {"label": Hamiltonian_Op[i][1], "coeff": {"real": Weight[i], "imag": 0.0}}
    Dict["paulis"].append(a)
    List_ops.append((Hamiltonian_Op[i][1], Weight[i]))

qo = PauliSumOp.from_list(List_ops)
Sum_Op = AbelianGrouper.group_subops(qo)
```

```python
# Here we are repartitioning the Hamiltonian into commuting group of Paulis, for later post-selection.

Hamiltonian_Op_IZ = []
Weight_IZ = []
Coeff_IZ = {}
f = open('QuHamiltonian.txt',"r")

for line in f:
    a = line.strip().split('    ')
    b = {str(a[0]): float(a[1])}
    c = (float(a[1]), str(a[0]))
    
    if 'X' not in str(a[0]) and 'Y' not in str(a[0]):
        Hamiltonian_Op.append(c)
        Hamiltonian_Op_IZ.append(c)
        Weight.append(float(a[1]))
        Weight_IZ.append(float(a[1]))
        Coeff.update(b)
        Coeff_IZ.update(b)
f.close()

List_IZ = []
for i in range(len(Weight_IZ)):
    List_IZ.append((Hamiltonian_Op_IZ[i][1], Weight_IZ[i]))
qo_IZ = PauliSumOp.from_list(List_IZ)
Sum_Op_IZ = AbelianGrouper.group_subops(qo_IZ)

Sum_Op += Sum_Op_IZ
```

```python
# Here we define a dictionary to store all the expectation values of each Pauli term with/without post-selection

ZEROS = np.zeros(len(Measured_List))
Expectation_Val_1 = dict(zip(Measured_List, ZEROS))
Expectation_Val_1_IZ = dict(zip(Measured_List, ZEROS))

for i in Measured_List:
    vars()[i] = 0.0
```

```python
#Circuit construction
def ansatz_circuit(params):
    wfc = QuantumCircuit(4)
    #changing the basis
    wfc.x(0)
    wfc.x(1)
    wfc.x(2)
    wfc.barrier(range(4))
    
    wfc.rx(-np.pi/2,1)
    wfc.ry(np.pi/2,3)
    wfc.cx(1,3)
    wfc.rz(params[0],3)
    wfc.cx(1,3)
    wfc.rx(np.pi/2,1)
    wfc.ry(-np.pi/2,3)
    
    return wfc
```

```python
qc = ansatz_circuit([0.10])
```

```python
def vqe(parameter):
    
    shots = 8192
    # Adding measurement error correction to the simulation
#     qr = QuantumRegister(4)
#     meas_calibs, state_labels = complete_meas_cal(qr=qr, circlabel='mcal')
#     job = execute(meas_calibs, backend=backend, shots=shots, coupling_map=coupling_map,
#                   noise_model=noise_model, initial_layout=[4,5,1,6])
#     cal_results = job.result()
#     meas_fitter = CompleteMeasFitter(cal_results, state_labels, circlabel='mcal')
#     meas_filter = meas_fitter.filter
    
    # Here we add all the circuits to be executed to a list to run
    qc_List = []
    for i in range(len(Sum_Op)):
        qc_intermediate = ansatz_circuit(parameter) + MeasureCircuit(Sum_Op[i])
        qc_List.append(qc_intermediate)
    job = execute(qc_List, backend=backend, shots=shots, coupling_map=coupling_map, noise_model=noise_model,
                  initial_layout=[4,5,1,6], optimization_level=0)
    result = job.result()
#     mitigated_result = meas_filter.apply(result)

#     # Apply the measurement error correction
#     for i in range(len(Sum_Op) - 1):
#         counts = mitigated_result.get_counts(qc_List[i])
#         for j in range(len(Sum_Op[i])):
#             vars()[Sum_Op[i][j].primitive.to_label()] = evaluation(counts, shots, Sum_Op[i][j].primitive.to_label())
    
    # For the last group in Sum_Op, it is the diagonal group and here we apply post-selection by using evaluation_IZ.
    counts = mitigated_result.get_counts(qc_List[-1])
    for j in range(len(Sum_Op[-1])):
        vars()[Sum_Op[-1][j].primitive.to_label()], effective_shots = evaluation_IZ(counts,
                                                                        Sum_Op[-1][j].primitive.to_label())    
    for i in Expectation_Val_1_IZ:
        if eval(i) != 0:
            Expectation_Val_1_IZ.update({i: eval(i)})
        else:
            None
    
    # This is without post-selection.
    for j in range(len(Sum_Op[-1])):
        vars()[Sum_Op[-1][j].primitive.to_label()] = evaluation(counts, shots, Sum_Op[-1][j].primitive.to_label())
                
    for i in Expectation_Val_1:
        if eval(i) != 0:
            Expectation_Val_1.update({i: eval(i)})
        else:
            None
    
    # H_expect_p is with post-selection
    H_expect_p = 0
    for i in range(len(Hamiltonian_Op)):
        H_expect_p += Hamiltonian_Op[i][0]*Expectation_Val_1_IZ.get(Hamiltonian_Op[i][1])
        
    H_expect = 0
    for i in range(len(Hamiltonian_Op)):
        H_expect += Hamiltonian_Op[i][0]*Expectation_Val_1.get(Hamiltonian_Op[i][1])   
    
    Variance = 0
    for i in range(len(Hamiltonian_Op)):
        if Expectation_Val_1_IZ.get(Hamiltonian_Op[i][1]) != 0:
            if 'X' in Hamiltonian_Op[i][1] or 'Y' in Hamiltonian_Op[i][1]:
                Variance += Hamiltonian_Op[i][0]**2*(1-Expectation_Val_1_IZ.get(Hamiltonian_Op[i][1])**2) / shots
            else:
                Variance += Hamiltonian_Op[i][0]**2*(1-Expectation_Val_1_IZ.get(Hamiltonian_Op[i][1])**2) / effective_shots
        else:
            Variance += 0
    stdev_p = np.sqrt(Variance)
    
    Variance = 0
    for i in range(len(Hamiltonian_Op)):
        if Expectation_Val_1.get(Hamiltonian_Op[i][1]) != 0:
            Variance += Hamiltonian_Op[i][0]**2*(1-Expectation_Val_1.get(Hamiltonian_Op[i][1])**2)
        else:
            Variance += 0
    stdev = np.sqrt(Variance/shots)
    
    print(parameter, H_expect, stdev)
    param.append(parameter[0])
    energy.append(H_expect)
    energy_p.append(H_expect_p)
    std.append(stdev)
    std_p.append(stdev_p)
    
    # For the vqe process, we will return the measured energy with post-selection to the optimizer
    return H_expect_p
```

```python
param = []
energy = []
energy_p = []
std = []
std_p = []

initial_point = [0.0]
optimizer = SPSA(maxiter=100)
vqe_result = optimizer.optimize(num_vars=1, objective_function=vqe, initial_point=initial_point)
print(vqe_result)
```
