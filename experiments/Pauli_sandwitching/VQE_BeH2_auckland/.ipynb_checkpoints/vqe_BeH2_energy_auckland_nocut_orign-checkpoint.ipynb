```python
def apply_checking_circuit(qc, ctrl_bits, ancilla_bits, side = None):
    if len(ctrl_bits) != len(ancilla_bits):
        print("Size mismatch")
        return None
    if side == 'front':
        for i in ancilla_bits:
            qc.h(i)
        for j,k in zip(ctrl_bits, ancilla_bits):
            qc.cz(j, k)
    elif side == 'end':
        for j,k in zip(ctrl_bits, ancilla_bits):
            qc.cz(j, k)
        for i in ancilla_bits:
            qc.h(i)
    else:
        print("Side undefined")
```

```python
pauli_check_circs = []
for i in range(0, 6):
    circuit = front_layer.copy()
    apply_checking_circuit(circuit, [i], [6], side = 'front')
    circuit += mid_layer
    apply_checking_circuit(circuit, [i], [6], side = 'end')
    circuit += end_layer
    pauli_check_circs.append(circuit)
```
